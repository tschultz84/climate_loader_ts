```python
#Create a functin to test if a point is inside the states file.

p1 = gdf['geometry'].iloc[1]

#This class checks whether a Point(lon,lat) is inside the USA.
class USA_check :
    def __init__(self):
        
        #First, load the USA SHapefile. Then, limit it to include geometries inside the contiguous 48 states.
        self.states = geopandas.read_file('C:\\Users\\14154\\OneDrive\\Python\\us_climate_analyzer\\data\\shapefiles\\cb_2018_us_state_20m.shp')
        self.states = self.states[self.states['NAME']!="Hawaii"]
        self.states = self.states[self.states['NAME']!="Puerto Rico"]
        self.states = self.states[self.states['NAME']!="Alaska"]
        
    def check(self,pt):
        #Then, see if the point is inside the shapefile. 
        poly=self.states
        out=  poly.contains(pt)
        #If yes, then there is at leaset one instance of true. 
        returner = out[out==True].count()
    
        return (returner>0)
        
print(USA_check().check(Point(-111,45)))
USA_test = USA_check()
    
```

### First, a Mesh of Points with Equal Separation between Lat / Lon
First, create a mesh equally spaced along latitude and longitude. I will show that this results in density that is not even across the northern and southern parts of the states. 

```python
#Find the bounds of the 48 contiguous US states.
lon_min=min(states.bounds['minx'])
lon_max=max(states.bounds['maxx'])
lat_min=min(states.bounds['miny'])
lat_max=max(states.bounds['maxy'])

lat_range = np.arange(lat_min,lat_max,1)
lon_range = np.arange(lon_min,lon_max,1)
#Use this to create a mesh.
meshlist = np.transpose(np.meshgrid(lat_range,lon_range)).reshape(-1,2)
#THen, limit this mesh only to points inside the contiguous USA.
sample = []
counter=0
interim=0
for i in meshlist:
    counter = counter + 1
    pt = Point(i[1],i[0])
    if USA_test.check(pt) == True:
        sample = sample+[i]
    if counter == 50:
        #print(f"Completed checking {interim} stations out of {len(meshlist)}.")
        interim=interim+counter
        counter=0
even_latlon_sample=np.asarray(sample).reshape(-1,2)
print(f"The dataset of the entire lower 48 states includes {len(usa_stations)} stations.")
print(f"The points to be searched includes {len(even_latlon_sample)} stations.")
```

# Create a Mesh Spaced on Distance

Here I create a mesh spaced purely on miles between each station. This spacing is calculated by using the Haversine equation.

I will then evaluate the station density it picks up, i.e., how far away each station is from each reference point, to help decide which to go with. 


```python
#Load Haversine equation.
os.chdir('C:\\Users\\14154\\OneDrive\\Python\\us_climate_analyzer\\')


#Example of using the inverse haversine.
paris = (48.8567, 2.3508) # (lat, lon)
# Finding 32 km west of Paris
mi = 300
km = mi/0.621371192

t1=inverse_haversine(paris, mi, Direction.WEST,unit=Unit.MILES)

print(f"Paris is located at {paris}, the location {mi} miles west is at {t1}")
```

### Create a Mesh with 100 mile spacing

```python
#This creates a class to make it so I can flexibly create a mesh with any spacing.
class SpaceMesh:
    def __init__(self,miles_between):
        self.miles_btwn=miles_between
        #First, load the USA SHapefile. Then, limit it to include geometries inside the contiguous 48 states.
        self.states = geopandas.read_file('C:\\Users\\14154\\OneDrive\\Python\\us_climate_analyzer\\data\\shapefiles\\cb_2018_us_state_20m.shp')
        self.states = self.states[self.states['NAME']!="Hawaii"]
        self.states = self.states[self.states['NAME']!="Puerto Rico"]
        self.states = self.states[self.states['NAME']!="Alaska"]
        
        #Find the bounds of the 48 contiguous US states.
        self.lon_min=min(self.states.bounds['minx'])
        self.lon_max=max(self.states.bounds['maxx'])
        self.lat_min=min(self.states.bounds['miny'])
        self.lat_max=max(self.states.bounds['maxy'])
        
        self.USA_test = USA_check()
    
    #miles_btwn is the miles between points.
    #limit is the maximum number of iterations.
    def create(self,limit=10000):
        #Initiliaze the points.
        sample1 = []

        #This initializes the loop.
        counter = 0
        later = self.lat_min
        loner = self.lon_min
        
        while counter < limit:
            counter = counter + 1
            #Create the intitial point.
            pt = Point(loner,later)
            if self.USA_test.check(pt) == True:
                #Updates the output only if the point is in the USA>
                sample1 = sample1+[later,loner]
            #Finds the new point.
            newpt=inverse_haversine((later,loner),self.miles_btwn, Direction.EAST,unit=Unit.MILES)
            #Assigns new coordinates. 
            later = newpt[0]
            loner=newpt[1]
            #If longitude goes past the maximum, then go to the next row northwards.
            if loner > self.lon_max:
                newpt=inverse_haversine((later,self.lon_min),self.miles_btwn, Direction.NORTH,unit=Unit.MILES)
                later = newpt[0]
                loner=newpt[1]
                #If the latitude is now also over latitude maximum, then you are done. 
                if later > self.lat_max:
                    print(f"Analysis done for a point spacing of {self.miles_btwn} miles! I generated a list of points, covering {len(sample1)/2} points.")
                    counter = limit + 1
        points_list = np.asarray(sample1).reshape(-1,2) 
        self.points_list=points_list
        return(points_list)
```

## Reviewing Average Distance from Mesh point to Weather Stations
I will now review the actual distance from each grid point to the weather station, to see what kind of mesh is useful. 

If a mesh is too fine, weather statinos will be far apart and there will be reundant weather stations. If too coarse, we miss data. 

```python
#This will review the distance between all points in each mesh.
mesh_list=[mesh200,mesh500]
recentyear=2020
stations_recent = contig_usa_stations[contig_usa_stations['Lastyear']>recentyear]
np_stations = np.asarray(stations_recent[['Latitude','Longitude']])
print(f"There are {len(stations_recent)} stations with the oldest data before 1920 and most recent after {recentyear} ")
print(np_stations)
```

```python
#I run the test over several different spacing options. 
dist_list = np.arange(25,500,25)
print(f"Searching across meshes of these mile spacing: {dist_list}")

#This is the empty results list. 
result_list = pd.DataFrame(columns={
    "Mesh Spacing (mi)",
    "Number of Mesh Points",
    #"Number of Weather Stations Selected",
    "Number of Redundantly Selected Stations",
    "Average Distance (mi) from Station to Mesh point",
    "90% Percentile Distance (mi) from Station to Mesh point",
    "Maximum  Distance (mi) from Station to Mesh point",
    "Average Distance (mi) to Next Closest Station"})

#Run over mesh.
#for mesh in mesh_list:
for dist in dist_list:
    sminit=SpaceMesh(dist)
    mesh = sminit.create()
    miles_result=[]
    stations_result=[]
    for i in mesh:
        #Calculate distances from mesh point i to every single point.
        interdist=[]
        pts_new=[]
        for j in np_stations:
            pt1 = i
            pt2 = j
            pts_new_int = haversine(pt1,pt2,unit=Unit.MILES)
            interdist = interdist + [pts_new_int]
            #pts_new = pts_new+[pts_new_int]
        #Find the minimum values. 
        min_loc = np.argmin(interdist)
        
        #Select only the closest station. 
        miles_result = miles_result + [interdist[min_loc]]
        stations_result = stations_result+[np_stations[min_loc]]
    
    #Now search the distance between stations.
    miles_between_sel_stations = []
    
    for k in stations_result:
        miles_between_chosen_stations = []
        for b in stations_result:
            miles_between_chosen_stations = miles_between_chosen_stations + [haversine(k,b,unit=Unit.MILES)]
        #Choosing the second closest station, since the very closest will alway shave zero dsitance
        #(i.e. this will be itself. )
        closest=np.sort(np.asarray(miles_between_chosen_stations))[1]
        miles_between_sel_stations = miles_between_sel_stations + [closest]
        
    
    #Take the mean distance between each mesh point and each grid point.
    av_miles = np.mean(miles_result)
    miles_max = np.max(miles_result)
    miles_90 = np.percentile(miles_result,90)
    
    #Take the mean distance between each chosen station.
    av_station_miles = np.mean(np.asarray(miles_between_sel_stations))
    
    #Then count the unique values. 
    count_unique = np.shape(np.unique(stations_result,axis=0))[0]
    #The difference between the total number of results and the unique values is the stations
    #which were found moer than once in a search. 
    noresults = len(stations_result)
    redundant = noresults - count_unique  
    
    result_list = result_list.append(
        {"Mesh Spacing (mi)": dist,
            "Average Distance (mi) from Station to Mesh point":av_miles,
             "90% Percentile Distance (mi) from Station to Mesh point":miles_90,
             "Maximum  Distance (mi) from Station to Mesh point":miles_max,
            "Number of Redundantly Selected Stations":redundant,
             "Number of Mesh Points": len(mesh),
             #"Number of Weather Stations Selected": noresults,
             "Average Distance (mi) to Next Closest Station": av_station_miles
            }, ignore_index=True)
    
    
```
